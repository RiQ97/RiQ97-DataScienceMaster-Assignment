### Q1: What is Estimation Statistics? Explain point estimate and interval estimate. 

A1. **Estimation Statistics** involves using sample data to estimate population parameters. It provides a way to infer properties of a larger population based on a smaller sample.

**Point Estimate** is a single value derived from sample data that serves as an estimate of a population parameter.
- **Example**: If we calculate the average height from a sample of 50 people, that average height is a point estimate of the population mean height.

**Interval Estimate** provides a range of values within which the population parameter is expected to lie, with a certain level of confidence.
- **Example**: If we have a 95% confidence interval for the average height of 150 cm to 160 cm, this means we are 95% confident that the population mean height falls within this range.

### Q2: Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
#A2.
def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level):
    import scipy.stats as stats

    # Calculate the critical value for the given confidence level
    alpha = 1 - confidence_level
    critical_value = stats.norm.ppf(1 - alpha / 2)
    
    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std_dev / (sample_size ** 0.5))
    
    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

### Q5: Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.


```python

#A5. 
import numpy as np
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    # Calculate sample means and standard deviations
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    std_dev1, std_dev2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)
    n1, n2 = len(sample1), len(sample2)

    # Calculate the t-statistic
    pooled_std_dev = np.sqrt((std_dev1**2 / n1) + (std_dev2**2 / n2))
    t_statistic = (mean1 - mean2) / pooled_std_dev

    # Calculate the degrees of freedom
    df = min(n1 - 1, n2 - 1)

    # Calculate the p-value
    p_value = 1 - stats.t.cdf(t_statistic, df)

    return t_statistic, p_value

# Example usage
sample1 = [60, 62, 61, 65, 64, 66]
sample2 = [55, 58, 57, 60, 59, 61]

t_statistic, p_value = two_sample_t_test(sample1, sample2)
print(f"T-statistic: {t_statistic:.2f}, P-value: {p_value:.4f}")
```

### Q6: What is a null and alternative hypothesis? Give some examples.

- **Example**: \( H_0 \): The mean test score of students who studied with method A is equal to the mean test score of students who studied with method B.

**Alternative Hypothesis (H1)** is a statement that contradicts the null hypothesis and represents the effect or difference that is being tested.

### Q7: Write down the steps involved in hypothesis testing.

A7. The steps involved in hypothesis testing are:
1. **State the Hypotheses**: Define the null and alternative hypotheses.
2. **Select the Significance Level**: Choose a significance level (α), commonly 0.05 or 0.01.
3. **Collect Data**: Gather and prepare the sample data.
4. **Perform the Test**: Use an appropriate statistical test to analyze the data.
5. **Calculate the Test Statistic and P-value**: Compute the test statistic and p-value based on the data and test.
6. **Make a Decision**: Compare the p-value to the significance level. If p≤α, reject the null hypothesis.
7. **Draw a Conclusion**: Interpret the results and make a conclusion based on the test.

### Q8: Define p-value and explain its significance in hypothesis testing.

A8. **P-value** is the probability of observing a test statistic at least as extreme as the one observed, assuming that the null hypothesis is true.

Significance of p-value:
- The p-value helps determine whether the observed data falls within the range of what we would expect under the null hypothesis. 
- A low p-value (below the significance level) indicates that the observed effect is unlikely to have occurred under the null hypothesis, leading to the rejection of the null hypothesis.

### Q10: Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.


```python

#A10. 
import numpy as np
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    # Calculate means, standard deviations, and sizes
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    std_dev1, std_dev2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)
    n1, n2 = len(sample1), len(sample2)

    # Calculate the t-statistic and degrees of freedom
    pooled_std_dev = np.sqrt((std_dev1**2 / n1) + (std_dev2**2 / n2))
    t_statistic = (mean1 - mean2) / pooled_std_dev
    df = min(n1 - 1, n2 - 1)  # Approximating degrees of freedom

    # Calculate the p-value
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))

    return t_statistic, p_value

# Example usage
sample1 = [85, 90, 88, 92, 87]
sample2 = [80, 85, 82, 86, 84]

t_statistic, p_value = two_sample_t_test(sample1, sample2)
print(f"T-statistic: {t_statistic:.2f}, P-value: {p_value:.4f}")
```

### Q12: What is t-statistic? State the formula for t-statistic.

A12. **T-Statistic** is a measure used in hypothesis testing that compares the difference between the sample mean and the population mean, scaled by the sample standard deviation and sample size.

For a single sample:
$$
t = \frac{\bar{X} - \mu} {\frac{s}{\sqrt{n}}}
$$
where \( \bar{X} \) is the sample mean, \( \mu \) is the population mean, \( s \) is the sample standard deviation, and \( n \) is the sample size.

For two independent samples:
$$
t = \frac{\bar{X_1} - \bar{X_2}}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$
where $$ \bar{X_1} $$ and $$ \bar{X_2} $$ are the sample means, $$ s_1^2 $$ and $$ s_2^2 $$ are the sample variances, and $$ n_1 $$ and $$ n_2 $$ are the sample sizes.

### Q13: A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.

A13. 
1. **Sample Mean (M)**: $500
2. **Sample Standard Deviation (s)**: $50
3. **Sample Size (n)**: 50
4. **Confidence Level**: 95% (Critical value for 95% confidence is approximately 1.96)

Calculate the margin of error:
$$
\text{Margin of Error} = 1.96 \times \frac{50}{\sqrt{50}}
$$

Confidence interval:
$$
\text{CI} = 500 \pm \text{Margin of Error}
$$



```python

import scipy.stats as stats
import numpy as np

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level):
    alpha = 1 - confidence_level
    critical_value = stats.norm.ppf(1 - alpha / 2)
    margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example data
sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)
print(f"95% Confidence Interval: (${lower_bound:.2f}, ${upper_bound:.2f})")
```

### Q14: A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

A14.
1. **Null Hypothesis (H0)**: The drug decreases blood pressure by 10 mmHg. $$ ( \mu = 10 ) $$
2. **Alternative Hypothesis (H1)**: The drug does not decrease blood pressure by 10 mmHg. $$ ( \mu \neq 10 ) $$
3. **Sample Mean (M)**: 8 mmHg
4. **Sample Standard Deviation (s)**: 3 mmHg
5. **Sample Size (n)**: 100
6. **Significance Level (α)**: 0.05

Calculate the t-statistic:
$$
t = \frac{8 - 10}{\frac{3}{\sqrt{100}}} = \frac{-2}{0.3} \approx -6.67
$$

Calculate the p-value:


```python

from scipy.stats import t

# Degrees of freedom
df = 100 - 1

# Calculate p-value
p_value = 2 * t.cdf(-abs(-6.67), df)
print(f"P-value: {p_value:.4f}")
```

### Q16: Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.

A16.
1. **Null Hypothesis (H0)**: The population means for the two groups are equal. (\( \mu_1 = \mu_2 \))
2. **Alternative Hypothesis (H1)**: The population means for the two groups are not equal. (\( \mu_1 \neq \mu_2 \))
3. **Group 1**: \( n_1 = 30 \), Mean = 80, SD = 10
4. **Group 2**: \( n_2 = 40 \), Mean = 75, SD = 8
5. **Significance Level (α)**: 0.01

Calculate the t-statistic:
$$
t = \frac{80 - 75}{\sqrt{\frac{10^2}{30} + \frac{8^2}{40}}}
$$

```python
mean1, std_dev1, n1 = 80, 10, 30
mean2, std_dev2, n2 = 75, 8, 40

t_statistic = (mean1 - mean2) / np.sqrt((std_dev1**2 / n1) + (std_dev2**2 / n2))
df = min(n1 - 1, n2 - 1)

p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
print(f"T-statistic: {t_statistic:.2f}, P-value: {p_value:.4f}")
```
