### Q1: Explain the following with an example:
1. Artificial Intelligence

2. Machine Learning

3. Deep Learning

A1. 
1. **Artificial Intelligence (AI):** refers to the broader concept of creating machines or systems that can perform tasks that typically require human intelligence. This includes tasks such as reasoning, problem-solving, understanding natural language, and perception.  
**Example:** A chatbot that can answer customer service questions by understanding and responding to user queries in natural language.

2. **Machine Learning (ML):** is a subset of AI that focuses on developing algorithms that allow computers to learn from and make predictions or decisions based on data. Unlike traditional programming, where the rules are explicitly coded, ML models learn from examples and improve over time.  
**Example:** An email spam filter that learns from labeled examples of spam and non-spam emails to predict whether new emails are spam.

3. **Deep Learning (DL):** is a subset of Machine Learning that involves neural networks with many layers (hence "deep"). These networks can model complex patterns in data and are particularly useful for tasks like image and speech recognition.  
**Example:** A deep learning model that can recognize and classify objects in images, such as identifying cats and dogs in photos.

### Q2: What is supervised learning? List some examples of supervised learning.

A2. **Supervised Learning:** is a type of machine learning where the model is trained on a labeled dataset. This means that each training example is paired with an output label, and the model learns to map inputs to the correct outputs.  

**Examples:**
**Classification:** Predicting whether an email is spam or not based on its content.
**Regression:** Predicting house prices based on features like size, location, and number of bedrooms.

### Q3: What is unsupervised learning? List some examples of unsupervised learning.

A3. **Unsupervised Learning:**  
Unsupervised Learning involves training a model on data without explicit labels. The goal is to identify patterns or structures within the data. The model tries to learn the underlying structure or distribution in the data.  

**Examples:**
**Clustering:** Grouping customers into segments based on purchasing behavior without predefined categories.
**Dimensionality Reduction:** Reducing the number of features in a dataset while retaining as much variance as possible, such as using Principal Component Analysis (PCA).

### Q4: What is the difference between AI, ML, DL, and DS?

A4.
**AI (Artificial Intelligence):** The broad field of creating intelligent agents or systems that can perform tasks requiring human-like intelligence.
**ML (Machine Learning):** A subset of AI focused on algorithms and models that learn from data and improve over time.
**DL (Deep Learning):** A subset of ML that uses deep neural networks with many layers to model complex patterns and relationships in data.
**DS (Data Science):** A broader field that combines statistics, data analysis, and machine learning to extract insights and knowledge from data. It encompasses both AI and ML, focusing on data-driven decision-making and problem-solving.

### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

A5.
**Supervised Learning:** Requires a labeled dataset where each input is associated with an output label. It is used for tasks like classification and regression.
**Unsupervised Learning:** Uses unlabeled data and aims to uncover hidden patterns or structures in the data. It is used for tasks like clustering and dimensionality reduction.
**Semi-Supervised Learning:** Combines a small amount of labeled data with a large amount of unlabeled data. It leverages the labeled data to guide the learning process, often improving performance when labeled data is scarce.

### Q6: What is train, test, and validation split? Explain the importance of each term.

A6.
**Train Split:** The portion of the dataset used to train the machine learning model. The model learns from this data by adjusting its parameters.
**Validation Split:** A portion of the dataset used to tune hyper-parameters and evaluate the model during training. It helps in selecting the best model and avoiding over-fitting.
**Test Split:** The portion of the dataset used to evaluate the final model's performance. It provides an estimate of how the model will perform on new, unseen data.

**Training Split:** Essential for teaching the model to learn patterns from data.
**Validation Split:** Helps in model selection and hyperparameter tuning without leaking information from the test set.
**Test Split:** Provides an unbiased evaluation of the model's performance and generalization ability.

### Q7: How can unsupervised learning be used in anomaly detection?

A7. Unsupervised learning can be used in anomaly detection by identifying data points that deviate significantly from the normal pattern of data. Techniques such as clustering (e.g., DBSCAN) or dimensionality reduction (e.g., PCA) can be employed to detect outliers or anomalies that do not conform to the learned patterns of the majority of the data.
