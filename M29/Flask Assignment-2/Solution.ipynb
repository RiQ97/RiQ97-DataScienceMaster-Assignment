Q1. Explain GET and POST methods.

A1. GET Method:

Used to request data from a specified resource.
Data is appended to the URL.
Can be cached.
Remains in the browser history.
Can be bookmarked.
Has length restrictions.
Should not be used when dealing with sensitive data.

POST Method:

Used to send data to a server to create/update a resource.
Data is sent in the request body.
Cannot be cached.
Does not remain in the browser history.
Cannot be bookmarked.
No restrictions on data length.
More secure for transmitting sensitive data.


Q2. Why is request used in Flask?

A2. In Flask, the request object is used to access incoming request data, such as form data, query parameters, headers, cookies, and files. It is essential for handling and processing the data sent by the client to the server.


Q3. Why is redirect() used in Flask?

A3. The redirect() function in Flask is used to redirect the client to a different endpoint or URL. 
It is useful when:
We want to direct the user to another page after a form submission.
We want to redirect to a login page if the user is not authenticated.
We want to ensure a clean URL structure by redirecting old URLs to new ones.


Q4. What are templates in Flask? Why is the render_template() function used?

A4. Templates are files that contain static data along with placeholders for dynamic content. They allow you to separate the presentation layer from the application logic.
Flask uses Jinja2 as its template engine, which provides a powerful way to generate HTML dynamically.

The render_template() function is used to render a template and return it as a response to the client. It takes the name of the template file and any number of keyword arguments to pass data to the template. This function helps in dynamically generating HTML content based on the data passed from the Flask application.


Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
#A5.

from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/api/greet', methods=['GET', 'POST'])
def greet():
    if request.method == 'POST':
        data = request.json
        name = data.get('name', 'World')
        return jsonify(message=f'Hello, {name}!')
    else:
        name = request.args.get('name', 'World')
        return jsonify(message=f'Hello, {name}!')

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
