### Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?


```python
#A1.
import pandas as pd

# Load the dataset
flight_data = pd.read_excel('flight_price.xlsx')

# Examine the dimensions
rows, columns = flight_data.shape
print(f"Rows: {rows}, Columns: {columns}")
```



### Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.


```python
#A2.
import matplotlib.pyplot as plt

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(flight_data['Price'], bins=30, edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```



### Q3. What is the range of prices in the dataset? What is the minimum and maximum price?


```python
#A3.

# Calculate minimum and maximum prices
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

print(f"Minimum Price: {min_price}, Maximum Price: {max_price}")
```


### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

A4.
1. Load the Dataset: Load the flight price dataset.
2. Clean and Prepare the Data: Ensure the data is in the correct format.
3. Create the Boxplot: Use seaborn or matplotlib to create a boxplot.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
flight_data = pd.read_excel('flight_price.xlsx')

# Display the first few rows to understand the structure of the dataset
print(flight_data.head())

# Create a boxplot to compare flight prices by airline
plt.figure(figsize=(14, 8))
sns.boxplot(x='Airline', y='Price', data=flight_data)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```


### Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

A5. The boxplot helps identify outliers as data points that fall outside the whiskers (1.5*IQR from the quartiles). Outliers can skew the analysis, affecting the mean and standard deviation, and may need to be treated or analyzed separately. 


```python
# Plot boxplot to identify outliers
plt.figure(figsize=(12, 8))
sns.boxplot(y='Price', data=flight_data)
plt.title('Boxplot of Flight Prices')
plt.ylabel('Price')
plt.show()
```


### Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?

A6. **Features to Analyze**:
1. **Date of Travel**: Extract month and year to identify seasonal patterns.
2. **Flight Demand**: Number of bookings over time.
3. **Price Trends**: Analyze price fluctuations over different times of the year.


```python
#1. Extract Month and Year

flight_data['Travel_Date'] = pd.to_datetime(flight_data['Date_of_Journey'], format='%d/%m/%Y')
flight_data['Travel_Month'] = flight_data['Travel_Date'].dt.month
flight_data['Travel_Year'] = flight_data['Travel_Date'].dt.year


#2. Visualize Seasonal Patterns

plt.figure(figsize=(14, 7))
sns.countplot(x='Travel_Month', data=flight_data)
plt.title('Flight Demand by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()

plt.figure(figsize=(14, 7))
sns.lineplot(x='Travel_Month', y='Price', data=flight_data, estimator='mean')
plt.title('Average Flight Prices by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.show()
```

### Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present your findings to your team?

A7. 
1. **Travel Date**: To observe trends over time.
2. **Airline**: To see if certain airlines have consistent pricing trends.
3. **Route**: Prices between specific origin-destination pairs.
4. **Class**: Economy, Business, First class prices.

Visualization will be like this :

```python
#1. Price Trends Over Time

plt.figure(figsize=(14, 7))
sns.lineplot(x='Travel_Date', y='Price', data=flight_data, estimator='mean')
plt.title('Trend of Flight Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()


#2. Price Trends by Airline

plt.figure(figsize=(14, 7))
sns.lineplot(x='Travel_Date', y='Price', hue='Airline', data=flight_data, estimator='mean')
plt.title('Flight Price Trends by Airline')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()


#3. Price Trends by Route

plt.figure(figsize=(14, 7))
sns.lineplot(x='Travel_Date', y='Price', hue='Route', data=flight_data, estimator='mean')
plt.title('Flight Price Trends by Route')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()
```
