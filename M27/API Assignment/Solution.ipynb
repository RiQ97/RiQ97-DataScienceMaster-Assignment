Q1. What is an API? Give an example, where an API is used in real life.

A1. An API is a set of rules and protocols for building and interacting with software applications. It allows different software systems to communicate with each other. APIs define the methods and data structures that developers can use to interact with a system, be it a library, application, or service.

Example: Social media integration on websites. For instance, when a website displays a feed of the latest tweets from Twitter or allows users to log in using their Facebook credentials, it uses the Twitter or Facebook API to fetch and display that information.


Q2. Give advantages and disadvantages of using API.

A2.
Advantages:

1. Integration: APIs allow different systems to integrate seamlessly, enabling data and functionality sharing.
2. Automation: APIs enable automation of tasks, improving efficiency and reducing manual work.
3. Scalability: APIs allow applications to scale by enabling modular architecture where different services can be updated or scaled independently.
4. Innovation: APIs enable developers to build new applications and services by leveraging existing systems and data.
5. Efficiency: APIs can provide specific, pre-defined functionalities that save time and effort for developers.

Disadvantages:

1. Security: APIs can introduce security vulnerabilities if not properly secured, as they expose endpoints that can be exploited.
2. Complexity: Designing and maintaining APIs can be complex, requiring careful planning and documentation.
3. Dependency: Applications relying on third-party APIs are dependent on the availability and performance of those APIs.
4. Compatibility: API changes can lead to compatibility issues, requiring updates to dependent applications.



Q3. What is a Web API? Differentiate between API and Web API.

A3. A Web API is a type of API that is accessible over the web using the HTTP/HTTPS protocol. It allows applications to communicate with each other over the internet.

Difference between API and Web API:

API: General term for any interface that allows interaction between software systems. It can be used locally within the same system or across different systems. It uses protocols like TCP, SMTP & HTTP.
Web API: Specifically refers to APIs that are accessible over the web using web protocols like HTTP/HTTPS. It typically involves communication between a client and a server.


Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

A4. REST is generally preferred for web and mobile applications due to its simplicity and performance, while SOAP is used in enterprise environments where advanced security and formal contracts are required.

REST (Representational State Transfer):

REST is an architectural style for designing networked applications. 
It uses stateless communication and relies on standard HTTP methods (GET, POST, PUT, DELETE).
Statelessness, client-server architecture, uniform interface, cacheability, layered system.
Typically uses JSON or XML for data exchange.


SOAP (Simple Object Access Protocol):

SOAP is a protocol for exchanging structured information in web services using XML.
It is highly extensible and supports various communication protocols like HTTP, SMTP, TCP.
Uses XML for message format.
Supports WS-Security for secure communication.

Shortcomings of SOAP:

1. Complexity: SOAP is more complex and requires more setup than REST.
2. Performance: SOAP messages are larger due to XML format, which can impact performance.
3. Scalability: SOAP’s stateful nature can make it less scalable compared to REST’s stateless architecture.
4. Flexibility: SOAP is less flexible and more rigid in terms of design compared to REST.
